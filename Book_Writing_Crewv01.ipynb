````python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.llms import Ollama
from typing import List, Any
import json, re, time
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm
import docx

# ──────────────────────────────────────────────────────────────────────────────
# 🔧 Configuration
# ──────────────────────────────────────────────────────────────────────────────
NUM_CH = int(input("🔢 Number of chapters? "))
initial_idea = input("🖋️ Seed idea? ")
book_title = input("📕 Book title? ")
mode = input("✍️ Mode? (philosophy/fiction): ").strip().lower()

# ──────────────────────────────────────────────────────────────────────────────
# 📚 Outline Prompt Templates
# ──────────────────────────────────────────────────────────────────────────────
outline_prompt_philosophy = PromptTemplate(
    input_variables=["topic", "num_chapters"],
    template="""
You are a philosophical and didactic writer. Your task is to generate a JSON array of exactly {num_chapters} chapter objects based on the following idea:

{topic}

Each chapter should reflect a key theme or concept that naturally arises from this idea. The style should resemble a contemplative guide or structured treatise (not a novel or fictional story).

Each chapter object should be in the following format:
{{
  "title": "...",         // A clear, concise, thematic title (e.g., "The Logic of Compassion")
  "description": "..."    // A one-sentence description of what the chapter explores or teaches
}}

Output ONLY a markdown-wrapped JSON array like this:
```json
[{{"title": "...", "description": "..."}}, ...]
```

Stay on topic. Avoid fiction. Focus on insight, self-mastery, philosophical depth, and logical principles.
"""
)

outline_prompt_fiction = PromptTemplate(
    input_variables=["topic", "num_chapters"],
    template="""
You are a creative fiction author. Generate a compelling outline of exactly {num_chapters} chapters for a fictional story based on the idea:

{topic}

Each chapter object should include a short, intriguing title and a one-sentence description summarizing the dramatic events or arc of that chapter.

Output ONLY a markdown-wrapped JSON array like this:
```json
[{{"title": "...", "description": "..."}}, ...]
```

Use clear structure and progression. Avoid overly generic titles.
"""
)

chapter_prompt = PromptTemplate(
    input_variables=["title", "description", "idea"],
    template="""
Write a chapter titled: {title}
This chapter is from a book based on the idea: {idea}

The chapter should follow this description:
{description}

Write about 4000–6000 words. Make it clear, logical, and well-organized.
Return plain text only.
"""
)

# ──────────────────────────────────────────────────────────────────────────────
# 🔗 Chain Setup
# ──────────────────────────────────────────────────────────────────────────────
llm = Ollama(model="llama3.1:latest")
outline_chain = LLMChain(
    prompt=outline_prompt_fiction if mode == "fiction" else outline_prompt_philosophy,
    llm=llm
)
chapter_chain = LLMChain(prompt=chapter_prompt, llm=llm)

# ──────────────────────────────────────────────────────────────────────────────
# 🧠 Outline Extraction
# ──────────────────────────────────────────────────────────────────────────────
def extract_array(raw) -> List[Any]:
    try:
        print("\n--- RAW OUTLINE RESPONSE ---\n", str(raw)[:1000], "...\n")
        if isinstance(raw, dict) and "text" in raw:
            raw = raw["text"]
        text = re.sub(r"<think>.*?</think>", "", raw, flags=re.S)
        m = re.search(r"```json\s*(\[[\s\S]*?\])\s*```", text)
        if not m:
            m = re.search(r'(\[[\s\S]*?\])', text)  # fallback
        if not m:
            raise ValueError("No JSON array found in output.")
        return json.loads(m.group(1))
    except Exception as e:
        print("❌ Error extracting chapters:", e)
        raise

# ──────────────────────────────────────────────────────────────────────────────
# 🚀 Generate Outline
# ──────────────────────────────────────────────────────────────────────────────
print(f"→ Generating “{book_title}” about:\n  {initial_idea}")
start = time.time()
raw_outline = outline_chain.invoke({"topic": initial_idea, "num_chapters": NUM_CH})
print(f"⏱️ Outline generation took {time.time() - start:.2f}s")
chapters = extract_array(raw_outline)
print(f"ℹ️ Outline returned {len(chapters)} chapters (target: {NUM_CH})")

# ──────────────────────────────────────────────────────────────────────────────
# 📝 Generate Chapters
# ──────────────────────────────────────────────────────────────────────────────
def safe_extract_text(result: Any) -> str:
    try:
        if isinstance(result, dict) and "text" in result:
            return result["text"]
        return str(result)
    except Exception as e:
        return f"[ERROR parsing result text: {e}]"

chapter_texts = [None] * len(chapters)
with ThreadPoolExecutor(max_workers=3) as executor:
    futures = {}
    for i, ch in enumerate(chapters):
        futures[executor.submit(chapter_chain.invoke, {
            "title": ch["title"],
            "description": ch["description"],
            "idea": initial_idea
        })] = i

    for fut in tqdm(as_completed(futures), total=len(futures), desc="Chapters"):
        i = futures[fut]
        try:
            result = fut.result()
            chapter_texts[i] = safe_extract_text(result)
        except Exception as e:
            chapter_texts[i] = f"[ERROR generating chapter {i+1}: {e}]"

# ──────────────────────────────────────────────────────────────────────────────
# 💾 Save to Word Doc
# ──────────────────────────────────────────────────────────────────────────────
doc = docx.Document()
doc.add_heading(book_title, 0)
doc.add_paragraph(f"Seed Idea: {initial_idea}")

for i, (meta, text) in enumerate(zip(chapters, chapter_texts)):
    doc.add_page_break()
    doc.add_heading(f"Chapter {i+1}: {meta['title']}", level=1)
    doc.add_paragraph(meta['description'], style='Intense Quote')
    doc.add_paragraph(text.strip() if isinstance(text, str) else str(text))

filename = f"{book_title.replace(' ', '_')}.docx"
doc.save(filename)
print(f"📘 Saved {filename}")
````
